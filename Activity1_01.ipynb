````python
import numpy as np
from random import choice

combo_indices = [['0,1,2'], ['0,3,6'], ['0,4,8'], ['3,4,5'], ['6,7,8'], ['1,4,7'], ['2,5,8'], ['2,4,6']]

EMPTY = "-"
IA = "X"
P = "O"

board = [[EMPTY,EMPTY,EMPTY],[EMPTY,EMPTY,EMPTY],[EMPTY,EMPTY,EMPTY]]

def boardP(board):
    print("\n")
    board_r = np.array(board)
    print(board_r)
    print("\n")
    
boardP(board)

def player_move(board, row, col):
    if board[row][col] != EMPTY:
        board[row][col] = P
        return board
    else:
        return "Espacio ocupado"
    

def all_moves_from_board(board, move_list):
    k = 0
    move_list = list()
    for i in range(3):
        for j  in range(3):
            if board[i][j] == EMPTY:
                move_list.append(str(k))
            k+=1
    return move_list

move_list = list()
move_list = all_moves_from_board(board, move_list)
print(move_list)

def ai_move(board, move_list):
    index = choice(move_list)
    
    if index == '0':
        board[0][0] = IA
    elif index == '1':
        board[0][1] = IA
    elif index == '2':
        board[0][2] = IA
    elif index == '3':
        board[1][0] = IA
    elif index == '4':
        board[1][1] = IA
    elif index == '5':
        board[1][2] = IA
    elif index == '6':
        board[2][0] = IA
    elif index == '7':
        board[2][1] = IA
    elif index == '8':
        board[2][2] = IA
    else:
        print("Next")

def game_won_by(board):
    won = False
    for i in combo_indices:
        if i == ['0,1,2']:
            for j in range(3):
                if board[j][0] == board[j][1] and board[j][1] == board[j][2] and board[j][2] != EMPTY:
                    print("Ganador: "+board[j][0])
                    won = True
                    break
        elif i == ['3,4,5']:
            for j in range(3):
                if board[1][j] == board[1][j] and board[1][j] == board[1][j] and board[1][j] != EMPTY:
                    print("Ganador: "+board[1][j])
                    won = True
                    break
        elif i == ['6,7,8']:
            for j in range(3):
                if board[2][j] == board[2][j] and board[2][j] == board[2][j] and board[2][j] != EMPTY:
                    print("Ganador: "+board[1][j])
                    won = True
                    break
        elif i == ['0,4,8']:
            if board[0][0] == board[1][1] and board[1][1] == board[2][2] and board[2][2] != EMPTY:
                print("Ganador: "+board[0][0])
                won = True
                break
        elif i == ['1,4,7']:
            if board[0][1] == board[1][1] and board[1][1] == board[2][1] and board[2][1] != EMPTY:
                print("Ganador: "+board[1][1])
                won = True
                break
        elif i == ['0,3,6']:
            if board[0][0] == board[1][0] and board[1][0] == board[2][0] and board[2][0] != EMPTY:
                print("Ganador: "+board[1][0])
                won = True
                break
        elif i == ['2,5,8']:
            if board[0][2] == board[1][2] and board[1][2] == board[2][2] and board[2][2] != EMPTY:
                print("Ganador: "+board[1][2])
                won = True
                break
        elif i == ['2,4,6']:
            if board[0][2] == board[1][1] and board[1][1] == board[2][0] and board[2][0] != EMPTY:
                print("Ganador: "+board[1][1])
                won = True
                break
        else:
            print("Next")
    
    return won
    
game_won_by(board)

def game_loop(board):
    board = [[EMPTY,EMPTY,EMPTY],[EMPTY,EMPTY,EMPTY],[EMPTY,EMPTY,EMPTY]]
    empty_cell_count = 9
    won = False
    while empty_cell_count > 0 and won == False:
        all_moves_from_board(board,move_list)
        if empty_cell_count % 2 == 1:
            ai_move(board, move_list)
        else:
            row = int(input('Fila: '))
            col = int(input('Columna: '))
            board = player_move(board, row, col)
            boardP(board)
            won = game_won_by(board)
                
    print("Juego terminado.")
````

As you can see, the tree of the board states consists of a total of $266073$ leaves. The `count_possibilities` function essentially implements a BFS algorithm to traverse all the possible states of the game. Notice that we count these states multiple times because placing an $X$ in the top-right corner in Step 1 and placing an $X$ in the top-left corner in Step 3 leads to similar possible states as starting with the top-left corner and then placing an $X$ in the top-right corner. If we implemented the detection of duplicate states, we would have to check fewer nodes. However, at this stage, due to the limited depth of the game, we will omit this step.

A decision tree, however, is identical to the data structure examined by `count_possibilities`. In a decision tree, we explore the utility of each move by investigating all possible future steps up to a certain extent. In our example, we could calculate the utility of the initial moves by observing the number of wins and losses after fixing the first few moves.
